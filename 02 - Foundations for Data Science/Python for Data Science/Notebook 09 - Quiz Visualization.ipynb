```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Count plot for the column 'Type'

```python
sns.countplot(data=df, x='Type')
```

### Strip plot of columns 'Rating' and 'Type'

```python
sns.stripplot(y = 'Rating', x='Type', hue='Type', data = df, legend=False)
```

### boxplot for the column 'Rating'm

```python
sns.boxplot(data=df, y='Rating')
```

### distribution of the column 'Reviews'

```python
sns.displot(df.Reviews, kind='kde')
```

### plot the heatmap of the correlation matrix of numerical variables

```python
sns.heatmap(df.corr(numeric_only=True))
```

```python
corr = df.corr(numeric_only=True)

# Plot the heatmap
plt.figure(figsize = (6, 3))
sns.heatmap(corr, annot = True, cmap = 'coolwarm',            
        fmt = ".1f",            
        xticklabels = corr.columns,            
        yticklabels = corr.columns)
```

```python
plt.scatter(df['Number_Of_Pages'],df['Price'])
plt.show()
```

### Price vs Rating

```python
plt.scatter(df['Price'],df['Rating'])
plt.show()
```

### pieplot for the column 'Type'

```python
plt.pie(df['Type'].value_counts(), autopct = '%.1f%%', radius = 1.2, 
        labels=list(df['Type'].value_counts().index))
plt.show()
```
