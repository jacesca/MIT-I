### 7. Why is stationarity important in time series analysis?

* It allows us to make accurate predictions about future values
* **It ensures that the mean, variance, and covariance of the data are constant over time** $\checkmark$
* It simplifies the process of data cleaning and preprocessing
* It allows us to use linear regression models to analyze the data

> A) Stationarity is important in time series analysis because it allows us to make predictions based on past patterns, but it does not guarantee that the predictions will be accurate. Other factors such as the choice of model, the quality of data, and the amount of noise in the data can also affect the accuracy of predictions.
> 
> B) Stationarity refers to the statistical properties of a time series that do not change over time, such as the mean, variance, and covariance. If a time series is stationary, it is easier to model and analyze, and the results of the analysis are more reliable.
> 
> C) While stationarity can make data cleaning and preprocessing easier in some cases, it is not always the case. In fact, some methods for achieving stationarity, such as differencing or detrending, can make the data more complex and difficult to work with.
>
> D) While stationarity is important for many time series analysis techniques, it is not a requirement for using linear regression models. Linear regression can be used on non-stationary data as well, but it may not be as effective or accurate as other methods.

### 8. What is the main difference between an AR model and an MA model?
* An MA model uses past observations to predict the current observation, while an AR model uses past errors.
* An AR model does not use past observations or past errors to predict the current observation.
* **An AR model uses past observations to predict the current observation, while an MA model uses past errors.** $\checkmark$
* An MA model does not use past observations or past errors to predict the current observation.

> An autoregressive (AR) model is a type of time series model that uses past observations to predict the current observation. Specifically, an AR model assumes that the current value of the time series is a linear combination of past values, with the weights of each past value determined by the autoregressive coefficient(s) of the model.
> 
> In contrast, a moving average (MA) model is a type of time series model that uses past errors to predict the current observation. Specifically, an MA model assumes that the current value of the time series is a linear combination of past errors, with the weights of each past error determined by the moving average coefficient(s) of the model.
>
> Therefore, the main difference between AR and MA models is the type of information they use to predict the current observation. AR models use past observations, while MA models use past errors.
